## What do we use **Statistics** for in Data Science ?

* To see summary(describe()) of our dataset
* Need to detect outliers in our dataset
* The check correlation in our dataset to avoid multicollinearity in our ML model
* To modify our existing probability
* To see the trend in our data

## How much Maths and Stats do you need to know ?

**The myth**
You need to have a very deep knowledge in mathematics and statistics in order to become a Data Scientist or Machine Learning Engineer

**The truth**
You need to have a good understanding of statistics and can confident in using the various statistical concepts

You are not a **Statistician** or a **Mathematician**

Our job is to have an understanding of the concepts and know how to use the tools based on the concepts to achieve results
